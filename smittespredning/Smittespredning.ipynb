Oppgave 1:

"I"=antall smittede
"a"==smitteraten

- Formelen blir da antall smittede pluss de nye 
    smittede
- Hvis man skal ha en generel oversikt overveksten av noe upåvirket av andre faktorer




```python
from pylab import*

I=3 #antall smittede
a=0.2 #smitterate
tid=48 #uker
populasjon= 157759 #total populasjon/ antall individer
tid_liste=[0]
smittet_liste=[I]


for i in range(tid):
    I=I+a*I*(1-(I/populasjon))
    tid_liste.append(i)
    smittet_liste.append(I)

plot(tid_liste, smittet_liste)
show()

```

Den er eksponensiell fordi vi ikke har noen begrensninger på hvor mange som kan bli smittet.

Det som skjer her er at ettersom  de smittede øker, blir det færre individer som er utsatte for å bli smittet. Siden de da blir imune.

```python
#Oppgave 3 og 4

from pylab import *

populasjon = 157759 # total populasjon / antall individer
I = 3 # Antall smittet
a = 4/populasjon # smitterate 
b = 3.5 # bedringsrate
S = populasjon - I  #De som ikke har blitt smittet / ikke er imune
R = 0 # Friske
tid = 48 # uker
v = 0.01 # Vaksinerate i prosent

tid_liste = [0]
utsatte_liste = [S]
smittet_liste = [I]
friske_liste = [R]

for i in range(1, tid):
    S = (S - a*I*S)*(1-v) 
    I = I + a*I*S - b*I
    R = R + b*I
    tid_liste.append(i)
    utsatte_liste.append(S)
    smittet_liste.append(I)
    friske_liste.append(R)

#plot(tid_liste, utsatte_liste, label = "utsatte")
plot(tid_liste, smittet_liste, label = "smittet")
#plot(tid_liste, friske_liste, label = "friske")
legend()
show()
```

her er en graf som fremstiller hvor mange som er smittete ved ulike tidspunkter. Grafen fremstiller antall smittede når vi har en vaksine med vaksineringsrate på 0.01 prosent.
