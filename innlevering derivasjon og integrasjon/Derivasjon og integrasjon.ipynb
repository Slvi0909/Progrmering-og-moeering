# Derrivasjon Oppgave 7.3 

```python
from pylab import*
e=2.718

def N(t):
    return (0.2*e**t)+90

def derrivert(N,t,dt):
    fder=(N(t+dt)-N(t))/dt
    return fder

antall=[]
tid=[]
endring=[]

for i in range(11):
    tid.append(i)
    antall.append(N(i)) 
    endring.append(derrivert(N,i,10**-8))   
       
plot(tid,antall)
plot(tid,endring)
legend(["planter","vekst"])
show()


```

# Integrasjon oppgave 7.14

```python
def f(x):
    return 2*x+1

def integralet(f,a,b,N):
    A=0
    dx = (b - a)/N
    for i in range(N):
        A=A+f(a+i*dx)*dx
    return A
print(integralet(f,1,4,3))
```

# integrasjon Oppgave 7.9 

```python
from pylab import*
e=2.718
tid=[]
temperatur=[]
endringtemperatur=[]
antall=[]

def T(t):
    return 70*(e**(-0.065*t))

def derrivert(T,t,dt):
    fder=(T(t+dt)-T(t))/dt
    return fder

for i in range(0,6001):
    x=i/1000
    tid.append(x)
    endringtemperatur.append(derrivert(T,x,10**-8))
    temperatur.append(T(x))

plot(tid,endringtemperatur)
plot(tid,temperatur)
legend(["temperaturendringer,""temperatur"])
show()

print("temperatur ved element 42 er",endringtemperatur[42])
    
```

# derrivasjon oppgave 7.12

```python
data=loadtxt("posisjonsdata",skiprows=1)
tid=data[:,0]
posisjon=data[:,1]

fart=[]
akselerasjon=[]

for i in range(len(tid)-1):
    dy=posisjon[i+1]-posisjon[i]
    dx=tid[i+1]-tid[i]
    der=dy/dx
    fart.append(der)

for i in range(len(tid)-2):
    dy=fart[i+1]-fart[i]
    dx=tid[i+1]-tid[i]
    der=dy/dx
    akselerasjon.append(der)
    

plot(tid,posisjon)
plot(tid[:-1],fart)
plot(tid[:-2],akselerasjon)

legend(["Posisjon","Fart","Akselerasjon"])
xlabel("Tid")
show()

```
